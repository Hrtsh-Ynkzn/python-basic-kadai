```python
# matplotlib.pyplotメソッドとnumpyをインポートするコードを記述する
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 2行3列のサブプロットを配置する
fig, axes = plt.subplots(2, 3)

# axes[0, 0]に折れ線グラフを描く
x0 = [0, 1, 2, 3]
y0 = [1, 2, -5, 2]
axes[0, 0].plot(x0, y0)

# axes[0, 1]にサインカーブを描く
x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)
axes[0, 1].plot(x1, y1)

# axes[0, 2]にヒストグラムを描く
x2 = np.random.normal(15, 5, 2000)
axes[0, 2].hist(x2, bins = 30)

# axes[1, 0]に散布図を描く
np.random.seed(3)
x3 = 4 + np.random.normal(0, 2, 60)        # 平均0, 標準偏差2で60要素を出力
y3 = 4 + np.random.normal(0, 3, len(x3))   # 平均0, 標準偏差3で(x3の要素数)個の要素を出力←個数をx3に依存させている
axes[1, 0].scatter(x3, y3)

plt.show()
```
